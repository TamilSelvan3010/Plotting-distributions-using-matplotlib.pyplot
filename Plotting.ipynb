```python
sea.distplot(diamond['price'],color='r')
mat.show()
```

```python
sea.boxplot(diamond['carat'])
mat.title('carat')
mat.show()
```

#### There are outliers high extreme value in the dataset in which plotted in 5 carat of diamond but the boxplot which has more data is plotted in the values of 0 to 1 carat of diamond but there are also outliers from 3 to 4 carat of diamond which is plotted in the dataset.

#### we have displayed the dataset in the vertical view in order to have clear view of understanding about dataset.

```python
sea.jointplot('carat','price',diamond,kind='reg',color='r')
mat.show()
```

#### In this jointplot we have Taken price and carat of the diamond ,in which 0.20 to 0.22 carat diamonds are sold at the price around 300 to 500 and 0.23 of carat diamonds are sold at the price of 300 to 700 ,0.24 carat of diamonds are being sold at the price of 350 to 750 and 0.25 carat of diamonds are sold at the price of 350 to 750, 0.26 carat of diamonds are sold at the price of  around 350 to 800, 0.27 to 0.29 carat of diamonds are being sold at price of 350 to 800 , and there is a outlier in 0.24 where it takes at the price of 900

```python
diamond=diamond[(diamond.carat<0.30)&(diamond.price<1000)]
sea.jointplot('carat','price',diamond,kind="resid",color='k')
mat.show()
```

#### in this jointplot we have chosen a certain value and condition for the two variables "carat" and "price" where price less 1000 and carat less than 0.30 in which there is a outlier plotted in price of 900 in 0.24 carat, in this condition there are average amount of value in 0.23 to 0.29 carat  but 0.20 to 0.22 carat has minimum value , the price range for the carat from 0.23 to 0.28 displays us it has the price range around 320 to 600 in an averge manner. 

```python
diamond=diamond[(diamond.carat<0.30)&(diamond.price<1000)]
sea.jointplot(x='carat',y='price',data=diamond,kind='hex',color='g')
mat.show()
```

#### we have displayed the variables of carat and price with a certain condition and also using color and kind where in this we have given color as red and kind as hexagon by using hexagon when the hexagon becomes darker it shows us there are maximum values are present in that if it becomes lighter it shows us there are minimum value are there, in this the most the dark hexagon is plotted in the value of (0.23,500) (carat,price) it also does displays the histograms of the respective variables.
